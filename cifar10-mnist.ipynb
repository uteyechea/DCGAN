```python
def set_dataset_shape(train_images, test_images):
    train_images = np.append(train_images, test_images, axis=0)
    if len(train_images.shape) == 3:
        train_images = train_images.reshape(train_images.shape[0],train_images.shape[1], train_images.shape[2], 1)
    return train_images

train_images = set_dataset_shape(train_images, test_images)
print(train_images.shape)
```

Get the attributes of the images

```python
def compute_min_image_resolution(sample_image):
    #Assuming all training images have the same resoltion
    height = sample_image.shape[0]
    width = sample_image.shape[1]
    
    while height % 2 == 0 and height > 4 and width % 2 == 0 and width > 4:
        height = height // 2
        width = width // 2
        
    return (height, width)


def compute_hidden_layers(low_resolution, sample_image):
    #Assuming all training images have the same resoltion
    height = low_resolution[0]
    width = low_resolution[1]
    
    max_height = sample_image.shape[0] 
    max_width = sample_image.shape[1]
    
    upscaling_y = []
    j = 0
    while height < max_height //2:
        j += 1
        height = height * 2
        upscaling_y.append(j)
    
    upscaling_x = []
    i = 0
    while width < max_width//2:
        i += 1
        width = width * 2
        upscaling_x.append(i)
    
    assert upscaling_x == upscaling_y
        
    return upscaling_x
```

```python
BUFFER_SIZE = train_images.shape[0]
HEIGHT = train_images.shape[1] 
assert HEIGHT % 4 == 0
WIDTH = train_images.shape[2] 
assert WIDTH % 4 == 0
CHANNELS = train_images.shape[3]
BATCH_SIZE = 128
EPOCHS = int(BUFFER_SIZE / BATCH_SIZE) * 5
IMG_SIZE=HEIGHT*WIDTH*CHANNELS
sample_image = train_images[0]# Test other source image resolution np.random.random((178,218)) #CelebA dataset
MIN_HEIGHT, MIN_WIDTH = compute_min_image_resolution(sample_image)
LAYERS = compute_hidden_layers((MIN_HEIGHT, MIN_WIDTH), sample_image)
```

## Create the models

Both the generator and discriminator are defined using the [Keras Sequential API](https://www.tensorflow.org/guide/keras#sequential_model).

### The Generator

The generator uses `tf.keras.layers.Conv2DTranspose` (upsampling) layers to produce an image from a seed (random noise). Start with a `Dense` layer that takes this seed as input, then upsample several times until you reach the desired image size of Width x Height x Channels. 

```python
def make_generator_model():
    
    filters = 128
    initializer = tf.keras.initializers.HeNormal()
    #initializer = 'glorot_uniform'
    layers_generator = LAYERS[::-1]
    
    model = tf.keras.Sequential()
    model.add(layers.Dense(MIN_HEIGHT * MIN_WIDTH * filters * max(layers_generator) * 2, kernel_initializer=initializer, use_bias=True, input_shape=(80,)))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU(0.2))
    model.add(layers.Dropout(0.5))

    model.add(layers.Reshape((MIN_HEIGHT, MIN_WIDTH, filters * max(layers_generator) * 2)))
    assert model.output_shape == (None, MIN_HEIGHT, MIN_WIDTH, filters * max(layers_generator) * 2)  # Note: None is the batch size

    for l in layers_generator:
        model.add(layers.Conv2DTranspose(filters*l, (5, 5), strides=(2, 2), padding='same', kernel_initializer=initializer, use_bias=False))
        model.add(layers.BatchNormalization())
        model.add(layers.LeakyReLU(0.2))
        model.add(layers.Dropout(0.5))
    
    model.add(layers.Conv2DTranspose(CHANNELS, (5, 5), strides=(2, 2), padding='same', kernel_initializer=initializer, use_bias=True, activation='tanh', kernel_regularizer='l2'))
    assert model.output_shape == (None, HEIGHT, WIDTH, CHANNELS)

    return model
```

Use the (as yet untrained) generator to create an image.

```python
generator = make_generator_model()
generator.summary()

noise = tf.random.normal([1, 80])

generated_image = generator(noise, training=False)
print(generated_image.shape)

plt.imshow((generated_image[0].numpy()*127.5+127.5).astype('int'))
```

### The Discriminator

The discriminator is a CNN-based image classifier.

Observations:
1. Filters and kernel size follow an inverse proportional relationship. Or remain constant.

2. A change in strides does not impact much the generated output.

3. Does the number of layers impact performance and output quality? an increase in the number of layers, decreases performance. The output image is better defined, as if it has a better control over the distribution of pixels.

4. Does the droput rate increase image quality? (droput rate could correlate with decreased creativity vs increased fidelity.)

```python
def make_discriminator_model():
    
    filters = 128
    initializer = tf.keras.initializers.HeNormal()
    #initializer = 'glorot_uniform'
    layers_discriminator = LAYERS
    
    model = tf.keras.Sequential()
    model.add(layers.Conv2D(filters, (5,5), strides=(2,2), padding='same', kernel_initializer=initializer, input_shape=[HEIGHT, WIDTH, CHANNELS]))
    #model.add(layers.BatchNormalization()) # Tips don't include BN over the first layer
    model.add(layers.LeakyReLU(0.2))
    #model.add(layers.Dropout(0.4))
    
    for l in layers_discriminator:
        model.add(layers.Conv2D(filters, (5,5), strides=(2,2), padding='same', kernel_initializer=initializer))
        model.add(layers.BatchNormalization())
        model.add(layers.LeakyReLU(0.2))
        #model.add(layers.Dropout(0.3))
    
    model.add(layers.Flatten())

    #model.add(layers.Dense(2*filters, kernel_initializer=initializer))
    #model.add(layers.BatchNormalization())
    #model.add(layers.PReLU())
    
    #model.add(layers.Dropout(0.1))
    
    model.add(layers.Dense(1, activation='sigmoid', kernel_initializer=initializer, use_bias = True))

    return model
```

```python
discriminator = make_discriminator_model()
discriminator.summary()
decision = discriminator(generated_image)
print(decision.shape)
print('Sample output: ', decision)
```

## Define the loss and optimizers

Define loss functions and optimizers for both models.


```python
# This method returns a helper function to compute cross entropy loss
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=False)
```

### Discriminator loss

This method quantifies how well the discriminator is able to distinguish real images from fakes. It compares the discriminator's predictions on real images to an array of 1s, and the discriminator's predictions on fake (generated) images to an array of 0s.

```python
def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss # Pending: Incorporate gradient penalty. Lambda gradient penalty factor.
    return total_loss
```

### Generator loss
The generator's loss quantifies how well it was able to trick the discriminator. Intuitively, if the generator is performing well, the discriminator will classify the fake images as real (or 1). Here, we will compare the discriminators decisions on the generated images to an array of 1s.

```python
def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)
```

The discriminator and the generator optimizers are different since we will train two networks separately.

```python
# Learning rate exponential decay
initial_learning_rate = 1e-3

lr_schedule_discriminator = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate,
    decay_steps=EPOCHS//1,
    decay_rate=0.99,
    staircase=False)
```

```python
# Learning rate exponential decay
initial_learning_rate = 1e-3

lr_schedule_generator = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate,
    decay_steps=EPOCHS//1,
    decay_rate=0.99,
    staircase=False)
```

```python
generator_optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule_discriminator, beta_1 = 0.0) #lr=1e-3
discriminator_optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule_generator, beta_1 = 0.0) #lr=1e-4
```

```python
#checkpoint_dir = os.path.join(data_dir,'training_checkpoints')
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt")
checkpoint = tf.train.Checkpoint(generator_optimizer=generator_optimizer,
                                 discriminator_optimizer=discriminator_optimizer,
                                 generator=generator,
                                 discriminator=discriminator)
```

## Define the training loop


```python

noise_dim = 80
num_examples_to_generate = 4*32

# We will reuse this seed overtime (so it's easier)
# to visualize progress in the animated GIF)
seed = tf.random.normal([num_examples_to_generate, noise_dim])
```

The training loop begins with generator receiving a random seed as input. That seed is used to produce an image. The discriminator is then used to classify real images (drawn from the training set) and fakes images (produced by the generator). The loss is calculated for each of these models, and the gradients are used to update the generator and discriminator.

```python
# Notice the use of `tf.function`
# This annotation causes the function to be "compiled".
@tf.function
def train_step(images):
    noise = tf.random.normal([BATCH_SIZE,noise_dim])
    
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
      generated_images = generator(noise, training=True)

      real_output = discriminator(images, training=True) 
      fake_output = discriminator(generated_images, training=True)

      gen_loss = generator_loss(fake_output)
      disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))

    return gen_loss,disc_loss
```

```python
history = []

def train(dataset, epochs):
  for epoch in range(epochs):
    start = time.time()

    for image_batch in dataset:
      gen_loss, disc_loss = train_step(image_batch)

    # Update the progress bar
    history.append([gen_loss.numpy().mean(),disc_loss.numpy().mean()])

    # Produce images for the GIF as we go
    display.clear_output(wait=True)
    generate_and_save_images(generator,
                             epoch + 1,
                             seed)
    # Visualize Generator & Discriminator Loss
    plotter(history,epoch)

    # Save the model every 15 epochs
    if (epoch + 1) % 15 == 0:
      checkpoint.save(file_prefix = checkpoint_prefix)

    print ('Time for epoch {} is {} sec'.format(epoch + 1, time.time()-start))

  # Generate after the final epoch
  display.clear_output(wait=True)
  generate_and_save_images(generator,
                           epochs,
                           seed)
  plotter(history,epoch)
```

**Train loss plotter**

```python
def plotter(history,epoch):
    x=np.arange(len(history))
    history=np.array(history)
    
    if len(history) < 10:
        gen_loss = history[:,0]
        disc_loss = history[:,1]
    else:
        gen_loss = history[10:,0]
        disc_loss = history[10:,1]
        x=x[10:]
        
    plt.figure(figsize=(24, 4)) 
    plt.title('Generator & Discriminator Loss')
    
    plt.subplot(1, 2, 1)
    plt.plot(x, gen_loss, label='Generator Loss')
    plt.legend(loc='upper left')
    
    plt.subplot(1, 2, 2)
    plt.plot(x, disc_loss, label='Discriminator Loss')
    plt.legend(loc='upper left')
    
    plt.show()
```

**Generate and save images**


```python
def generate_and_save_images(model, epoch, test_input):
  # Notice `training` is set to False.
  # This is so all layers run in inference mode (batchnorm).
  predictions = model(test_input, training=False)

  fig = plt.figure(figsize=(48,2*3))
  for i in range(predictions.shape[0]):
      plt.subplot(4, 32, i+1)
      plt.imshow((predictions[i].numpy()*127.5+127.5).astype('int') )
      plt.axis('off')

  plt.savefig('image_at_epoch_{:04d}.png'.format(epoch))
  plt.show()
```
